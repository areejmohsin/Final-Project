## Data Set 3 - Home Value

![home](https://media.giphy.com/media/kd9iaS9bRfjRPVy9dM/giphy.gif "home")

```python
import pandas as pd
import geopandas as gpd
from matplotlib import pyplot as plt

# read file in
price = pd.read_csv("Neighborhood.csv")


```

```python
sortedprice = price[['SNAP_All_csv_Neighborhood', 'SNAP_All_csv_Median_Home_Value_']].sort_values(by="SNAP_All_csv_Median_Home_Value_", ascending = False)


```

```python
# drops the naN values but doesn't apply it to the graph: 87 rows x 2 columns instead of 90 rows x 2 columns

sortedprice = sortedprice[['SNAP_All_csv_Neighborhood', 'SNAP_All_csv_Median_Home_Value_']].dropna()
```

```python
#get rid of the commas and dollars signs so data can be calculated
dollar = sortedprice.columns
sortedprice[dollar] = sortedprice[dollar].replace({'\$':''}, regex = True)
sortedprice[dollar] = sortedprice[dollar].replace({'\,':'.'}, regex = True)
# drops naN value and converts into float
sortedprice['SNAP_All_csv_Median_Home_Value_'].astype(float).dropna()
```

```python
#Median Home Value is now readable for python
sortedprice['SNAP_All_csv_Median_Home_Value_']=sortedprice['SNAP_All_csv_Median_Home_Value_'].astype(float)
sortedprice[['SNAP_All_csv_Neighborhood', 'SNAP_All_csv_Median_Home_Value_']].describe()
```

There are about 87 neighborhoods recorded here. The median home value max is 254k while the min is 17k. This data was taken from 2010 so the prices of homes don't fit the market today. 

```python
#let's get a better visual with a graph
sortedprice['SNAP_All_csv_Median_Home_Value_']=sortedprice['SNAP_All_csv_Median_Home_Value_'].astype(float)
graph = sortedprice.plot.barh(x="SNAP_All_csv_Neighborhood", y="SNAP_All_csv_Median_Home_Value_", figsize=(15,price.shape[0] * 0.3), legend=None)
graph.set_xlabel("Median Home Value")
```

```python

#let's shorten the graph to the higher median values 
graph = sortedprice[:15].plot.barh(x="SNAP_All_csv_Neighborhood",y="SNAP_All_csv_Median_Home_Value_",figsize=(15,sortedprice.shape[0] * 0.3), legend=None)
graph.set_xlabel("Median Home Value")
```

```python
#final graph presentation 
graph = sortedprice[:5].plot.barh(x="SNAP_All_csv_Neighborhood", y="SNAP_All_csv_Median_Home_Value_", figsize=(15,sortedprice.shape[0] * 0.3), legend=None)
graph.set_xlabel("Median Home Value")
```

As you can see the homes with the highest median value are:
1. Squirrel Hill South
2. North Oakland
3. Strip District 
4. Point Breeze
5. Allegheny West
