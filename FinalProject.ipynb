## Data Set 3 - Trees

```python
import pandas as pd
import geopandas as gpd
from matplotlib import pyplot as plt
from math import isnan
%matplotlib inline
```

```python
data = pd.read_csv("https://data.wprdc.org/datastore/dump/1515a93c-73e3-4425-9b35-1cd11b2196da", low_memory = False)
data.head(10)
```

```python
#This code checks to see if the street is in the dictionary. If it is, then I added one to the key. If it is not, or contains a string value, which makes it NaN, then I set it to one.
data_dict = dict()
for i in range(len(data)):
    street = data.loc[i, 'street']
    if street in data_dict:
        data_dict[street] += 1
    elif isinstance(street, str):
        data_dict[street] = 1
```

```python
#If a street has a value less than 300, then I removed it from the dictionary.
for street in list(data_dict.keys()):
    if data_dict[street] < 300:
        del data_dict[street]
#I then converted the dictionary back into a Series     
data_series = pd.Series(data_dict, dtype='float64')
data_series.sort_values(ascending = False)[:]
```

```python
data_series.sort_values(ascending = False)[:].plot.bar()
plt.title('Trees in Neighborhoods')
plt.xlabel('Neighborhoods')
plt.ylabel('Trees')
```

From this plot, we can see that Penn Ave has the most amount of trees, and is hence the best neighborhood. I have to do a lot more filtering to pinpoint the exact locations of the streets, but from looking at a map, the majority of these streets like Penn Ave and Forbes Ave go through Squirrel Hill. Thus, Squirrel Hill (including North and South) is the best neighborhood.
